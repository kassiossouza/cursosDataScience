**Desenvolvimento de uma função para a plotagem de gráficos nas análises feitas.**

```
def plotarGrafico(titulo, labelx, labely, x, y, dataset):
  sns.set_palette('Accent')
  sns.set_style('darkgrid')
  ax = sns.lineplot(x= x, y = y, data = dataset)
  ax.figure.set_size_inches(10, 5)
  ax.set_title(titulo, loc = 'left', fontsize = 16)
  ax.set_xlabel(labelx, fontsize = 13)
  ax.set_ylabel(labely, fontsize = 13)
  ax = ax
```

```
plotarGrafico('Aumento das Vendas da Alucar nos Anos de 2017 e 2018', 'Tempo', 'Aumento das Vendas (R$)', 'mes', 'aumento', alucar)
```

```
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```
plotarGrafico('Aceleração das Vendas da Alucar Mês a Mês Entre 2017 e 2018', 'Tempo', 'Aceleração', 'mes', 'aceleracao', alucar)
```

```
plt.figure(figsize = (14,10))
ax = plt.subplot(3, 1, 1)
ax.set_title('Análise das Vendas da Alucar de 2017 e 2018', fontsize = 18)
sns.lineplot(x = 'mes', y = 'vendas', data = alucar)
plt.subplot(3, 1, 2)
sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
plt.subplot(3, 1, 3)
sns.lineplot(x = 'mes', y = 'aceleracao', data = alucar)
ax = ax
```

```
def plotComparacao(x, y1, y2, y3, dataset, titulo):
  plt.figure(figsize = (14,10))
  ax = plt.subplot(3, 1, 1)
  ax.set_title(titulo, fontsize = 18)
  sns.lineplot(x = x, y = y1, data = dataset)
  plt.subplot(3, 1, 2)
  sns.lineplot(x = x, y = y2, data = dataset)
  plt.subplot(3, 1, 3)
  sns.lineplot(x = x, y = y3, data = dataset)
  ax = ax
```

**Gráficos Para a Análise das Vendas da Alucar ao Longo dos Anos de 2017 e 2018**

```
plotComparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'Análise das Vendas da Alucar em 2017 e 2018')
```

**Statsmodels**

```
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
resultado = seasonal_decompose([chocolura['vendas']], freq = 1)
ax = resultado.plot()
```

```
resultado = seasonal_decompose([chocolura['vendas']], freq = 2)
ax = resultado.plot()
```

```
resultado = seasonal_decompose([chocolura['vendas']], freq = 3)
ax = resultado.plot()
```

```
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```
data = ({
    'observacao' : observacao,
    'tendencia' : tendencia,
    'sazonalidade' : sazonalidade,
    'ruido' : ruido
})

resultado = pd.DataFrame(data)
resultado.head(8)
```

```
plotComparacao(resultado.index, 'observacao', 'tendencia', 'sazonalidade', resultado, 'Exemplo de Statsmodels')
```

```
plotComparacao(resultado.index, 'observacao', 'tendencia', 'ruido', resultado, 'Exemplo de Statsmodels')
```
