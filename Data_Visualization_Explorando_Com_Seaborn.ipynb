**Teste de hipótese**

**H<sup>null</sup>**

> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**

> **A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```
from scipy.stats import ranksums
```

```
gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```
comSobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```
comSobremesa
```

```
semSobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
semSobremesa
```

```
ranksums(comSobremesa, semSobremesa)
```

```
r = ranksums(comSobremesa, semSobremesa)
```

**H<sup>null</sup>**

> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

```
print('O valor do p-value é {}.'.format(r.pvalue))
```
