```python
#Criando funções para análise exploratória.
```

```python
def plotar_grafico_linha(titulo, xlabel, ylabel, x, y, dataset, hue):
    ax = sns.lineplot(x = x, y = y, hue = hue, data = dataset)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc = 'center', fontsize = 20)
    ax.set_xlabel(xlabel, fontsize = 20)
    ax.set_ylabel(ylabel, fontsize = 20, rotation = 90)
```

```python
def plotar_grafico_barra(titulo, xlabel, ylabel, x, y, dataset):
    plt.figure(figsize = (20, 6))
    ax = plt.bar(dataset[x], dataset[y])
    plt.title(titulo, loc = 'center', fontsize = 20)
    plt.xlabel(xlabel, fontsize = 20)
    plt.ylabel(ylabel, fontsize = 20, rotation = 90)
```

```python
plotar_grafico_barra("Novos Casos em SP", 'Data', 'Novos casos', 'data', 'novosCasos', df_sp)
```
